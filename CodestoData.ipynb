Read in USA 2017

```python
usa2017 = pd.read_csv('d3data/usa2017.csv')
```

```python
usa2017 = usa2017[(usa2017['p19_1l'] > 0) & (usa2017['p19_1l'] < 999999997)]
usa2017 = usa2017[(usa2017['p19_2l'] > 0) & (usa2017['p19_2l'] < 999999997)]
usa2017 = usa2017[(usa2017['p20l'] > 0) & (usa2017['p20l'] < 999999997)]
usa2017 = usa2017[(usa2017['p26'] > 0) & (usa2017['p26'] < 999999997)]
usa2017 = usa2017[(usa2017['p11_1d'] > 0) & (usa2017['p11_1d'] < 97)]
usa2017 = usa2017[(usa2017['p11_1m'] > 0) & (usa2017['p11_1m'] < 9997)]
usa2017 = usa2017[(usa2017['p14_1'] > 0) & (usa2017['p14_1'] < 9999997)]
```

```python
usa2017.to_csv('d3data/cutusa.csv', encoding='latin1')
```
