```python
import pandas as pd
import googlemaps
```

```python
usa = pd.read_csv('d3data/encodedusa.csv', encoding='latin1')
```

```python
usa_counts = usa[['first_citynew', 'second_citynew', 'long_citynew', 'us_entrynew']]
```

```python
counts = usa_counts.apply(pd.Series.value_counts)
counts.fillna(0)
counts['sum'] = counts.sum(axis=1)
```

```python
totals = counts.sort_values(by='sum')
```

```python
use = totals[['sum']]
use.reset_index(level=0, inplace=True)
use.head()
```

```python
mex_ports = usa[['mex_port', 'mex_port_lat', 'mex_port_long']][1131:1300]
```

```python
mex_ports['mex_port'].value_counts()
```

```python
mex_use = mex_ports.reset_index(drop=True)
mex_use.head()
```

```python
final  = mex_use.merge(use, left_index=True, right_index=True)
```
